# Basic Definitions


# One-One Function
$f:A→B$ is said to be one-to-one iff $(a1) =f(a2)$ forces $a1=a2$.

# Onto Function
$f:A→B$ is said to be onto if for any $b∈B$ there exists $a∈A$ such that $f(a)=b$

# Cardinalities of Sets
A set which is in bijection with some $J_{n}$ is called finite, one which is in bijection with $\mathbb{N}$ is called countable and a set which is neither finite nor countable is called uncountable.
